```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../churn_data/Bank_churn.csv")
```

```python
print("Dataset Overview:")
print(df.head())
print("\nSummary of Data Types and Missing Values:")
print(df.info())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='churn', data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['tenure'], kde=True, bins=20)
plt.title('Customer Tenure Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x='churn', y='balance', data=df)
plt.title('Balance by Churn Status')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='geography', hue='churn', data=df)  # Example of categorical feature
plt.title('Geography by Churn Status')
plt.show()
```
